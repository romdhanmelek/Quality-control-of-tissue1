```python
import cv2
from matplotlib import pyplot as plt
```

```python
scaleDown = 2
img = cv2.imread('test2.jpg', cv2.IMREAD_COLOR)  # lecture de l'image 
frameHeight, frameWidth, d = img.shape  # les dimensions de l'image
img = cv2.resize(img, (int(frameWidth/scaleDown), int(frameHeight/scaleDown)))  # réduire la taille de l'image

# Convert the image to gray-scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # convertir en niveau de gris
# Find the edges in the image using canny detector
plt.imshow(gray,cmap = 'gray')
plt.show()
```

```python
########### chercher les contours 
edges_cv = cv2.Canny(gray, 50, 255, 20)  # trouver les contours

plt.imshow(edges_cv,cmap = 'gray')
plt.show()
edges_cv.shape
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
#img = cv.imread('sudoku.png',0)
#img = cv.medianBlur(img,5)

scaleDown = 8
img_0 = cv2.imread('test2.jpg', cv2.IMREAD_COLOR)  # lecture de l'image 

frameHeight, frameWidth, d = img_0.shape  # les dimensions de l'image
img_1 = cv2.resize(img_0, (int(frameWidth/scaleDown), int(frameHeight/scaleDown)))  # réduire la taille de l'image

img = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
#img = cv2.GaussianBlur(img, (5, 5), 0)
#img =cv2.equalizeHist(img)
#img = cv2.GaussianBlur(img, (5, 5), 0)
th = 200


ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
